```python
import sys
import pandas as pd
import numpy as np
from pandas.tseries.offsets import BDay
```

```python
strt= "05/11/2018"
end = "21/12/2018"
project = 'ANAIS-2'
activity  = 'CU'
```

```python
d1 = pd.to_datetime(strt, format='%d/%m/%Y')
d2 = pd.to_datetime(end, format='%d/%m/%Y')
```

```python
dd = [d for d in pd.date_range(d1, d2).tolist() if d.dayofweek != 6]
```

```python
d1
```

```python
df = pd.DataFrame(pd.date_range(d1, d2), columns=['date'])
```

```python
df = df.set_index('date')
df = df[df.index.dayofweek < 5]
df['datef'] = df.index.strftime("%d/%m/%Y")
```

```python
df.index.dayofweek
```

```python
df['hours'] = np.where(df.index.dayofweek==4, "6,25", "8,75")
```

```python
df['project'] = project
df['activity'] = activity
```

```python
df.to_csv("hours.csv", sep=';', header=False, index=False)
```
